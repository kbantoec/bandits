# Experiment 1: Bandit program with epsilon-greedy

```python
from typing import List

from experiment import experiment
```

```python
bandit_probabilities: List[float] = [0.2, 0.5, 0.75]  # Probabilities for each bandit
epsilon: float = 0.1
num_trials: int = 10_000
```

```python
experiment(bandit_probabilities, num_trials, epsilon)
```

Observations:
* The optimal bandit is chosen correctly as `optimal_j` is at index 2
* Means are quite close to the true means which are 0.2, 0.5 and 0.75
* The total reward of 7,176 is close to the expected reward or 7,500 if we had selected the optimal bandit each time; of course we didn't as we allow exploration so our reward is naturally lower
* We can see that our ratio exploration / exploitation is approximately 1 to 10 as expected (epsilon = 0.1)
